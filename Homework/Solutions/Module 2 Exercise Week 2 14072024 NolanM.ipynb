```python
import numpy as np

def compute_vector_length(vector):
    np_vector = np.array(vector)
    # Compute the Euclidean norm (length) of the vector
    return np.linalg.norm(np_vector)

vector = [3, 4]
length = compute_vector_length(vector)
print(f"The length of the vector {vector} is {length}")
```

```python
def compute_dot_product(vector1, vector2):
    if not isinstance(vector1, (list, np.ndarray)) or not isinstance(vector2, (list, np.ndarray)):
        raise TypeError("Both inputs must be lists or NumPy arrays.")
    vector1 = np.array(vector1)
    vector2 = np.array(vector2)
    # Check if both vectors have the same length
    if vector1.shape != vector2.shape:
        raise ValueError("Both vectors must have the same length.")
    # Compute the dot product
    return np.dot(vector1, vector2)

vector1 = [1, 2, 3]
vector2 = [4, 5, 6]
result = compute_dot_product(vector1, vector2)
print(f"The dot product of {vector1} and {vector2} is {result}")
```

```python
def matrix_multi_vector(vector, matrix):
    if not isinstance(vector, (list, np.ndarray)):
        raise TypeError("The vector must be a list or a NumPy array.")
    if not isinstance(matrix, (list, np.ndarray)):
        raise TypeError("The matrix must be a list or a NumPy array.")
    
    vector = np.array(vector)
    matrix = np.array(matrix)
    
    # Check if the vector's length matches the matrix's number of rows
    if vector.shape[0] != matrix.shape[1]:
        raise ValueError("The vector's length must match the number of columns in the matrix.")
    return np.dot(vector, matrix)

vector = [1, 2, 3]
matrix = [
    [4, 5, 6],
    [7, 8, 9],
    [10, 11, 12]
]
result = matrix_multi_vector(vector, matrix)
print(f"The result of multiplying the vector {vector} by the matrix {matrix} is {result}")
```

```python
def matrix_multiplication(matrix1, matrix2, method='dot'):
    if not isinstance(matrix1, (list, np.ndarray)) or not isinstance(matrix2, (list, np.ndarray)):
        raise TypeError("Both inputs must be lists or NumPy arrays.")
    matrix1 = np.array(matrix1)
    matrix2 = np.array(matrix2)
    if matrix1.shape[1] != matrix2.shape[0]:
        raise ValueError("The number of columns in the first matrix must match the number of rows in the second matrix.")
    if method == 'dot':
        return np.dot(matrix1, matrix2)
    elif method =='matmul':
        return np.matmul(matrix1, matrix2)
    else:
        raise TypeError("The method is not available.")

matrix1 = [
    [1, 2, 3],
    [4, 5, 6]
]
matrix2 = [
    [7, 8],
    [9, 10],
    [11, 12]
]
result = matrix_multiplication(matrix1, matrix2, 'matmul')
print(f"The result of multiplying the matrices is:\n{result}")
```

```python
def inverse_matrix(A):
    # Check if the input is a square matrix
    if not isinstance(A, (list, np.ndarray)):
        raise ValueError("Input must be a list or a NumPy array.")
    A = np.array(A)
    if A.shape[0] != A.shape[1]:
        raise ValueError("Input must be a square matrix.")
    # Calculate the determinant of A
    det_A = np.linalg.det(A)
    if det_A == 0:
        return det_A, None  # Matrix is not invertible
    # Calculate the inverse of A
    inv_A = np.linalg.inv(A)
    return det_A, inv_A

A = [[-2, 6],
    [8, -4]]

det_A, inv_A = inverse_matrix(A)

print(f"Matrix A:\n{np.array(A)}")
print(f"Determinant of det A: {det_A}")
if inv_A is not None:
    print(f"Inverse of A:\n{inv_A}")
else:
    print("Matrix A is not invertible.")
```

```python
def eigen_decomposition(matrix):
    if not isinstance(matrix, (list, np.ndarray)):
        raise TypeError("Input must be a list or a NumPy array.")
    matrix = np.array(matrix)
    # Check if the matrix is square
    if matrix.shape[0] != matrix.shape[1]:
        raise ValueError("Input must be a square matrix.")
    # Calculate the eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eig(matrix)
    # Normalize the eigenvectors
    normalized_eigenvectors = eigenvectors / np.linalg.norm(eigenvectors, axis=0)
    return eigenvalues, eigenvectors, normalized_eigenvectors


A = [[0.9, 0.2],
    [0.1, 0.8]]

try:
    eigenvalues, eigenvectors, normalized_eigenvectors = eigen_decomposition(A)
    print("Matrix A:")
    print(np.array(A))
    print("\nEigenvalues:")
    print(eigenvalues)
    print("\nEigenvectors (unnormalized):")
    print(eigenvectors)
    print("\nEigenvectors (normalized):")
    print(normalized_eigenvectors)
except Exception as e:
    print(f"Error: {e}")
```

```python
def cosine_similarity(x, y):
    x = np.array(x)
    y = np.array(y)
    # Check if the vectors have the same size
    if x.shape != y.shape:
        raise ValueError("Vectors must have the same size.")
    # Calculate the dot product
    dot_product = np.dot(x, y)
    # Calculate the norms (magnitudes) of the vectors
    norm_x = np.linalg.norm(x)
    norm_y = np.linalg.norm(y)
    # Calculate the cosine similarity
    return dot_product / (norm_x * norm_y)

x = [1, 2, 3, 4]
y = [1, 0, 3, 0]

try:
    cosine_sim = cosine_similarity(x, y)
    print("Vector x:")
    print(np.array(x))
    print("\nVector y:")
    print(np.array(y))
    print("\nCosine Similarity:")
    print(cosine_sim)
except Exception as e:
    print(f"Error: {e}")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

class replace_background_module:
    def __init__(self, image_path, background_path):
        self.bg1_image = cv2.imread('./content/Week-2-Module-2/GreenBackground.png', 1)
        self.bg1_image = cv2.resize(self.bg1_image, (678, 381))

        self.obj_image = cv2.imread(image_path, 1)
        self.obj_image = cv2.resize(self.obj_image, (678, 381))

        self.bg2_image = cv2.imread(background_path, 1)
        self.bg2_image = cv2.resize(self.bg2_image, (678, 381))

        self.difference_single_channel = self.compute_different(self.bg1_image, self.obj_image)
        self.binary_mask = self.compute_binary_mask()
        self.result_image = self.apply_mask()

    def compute_different(self, bg_image, input_image):
        # Convert images to HSV color space
        bg_hsv = cv2.cvtColor(bg_image, cv2.COLOR_BGR2HSV)
        input_hsv = cv2.cvtColor(input_image, cv2.COLOR_BGR2HSV)
        # Compute the absolute difference in the HSV space
        difference = cv2.absdiff(bg_hsv, input_hsv)
        # Extract the V channel (brightness)
        return difference[:, :, 2]

    def compute_binary_mask(self):
        # Threshold the difference image to create a binary mask
        _, binary_mask = cv2.threshold(self.difference_single_channel, 50, 255, cv2.THRESH_BINARY)
        return binary_mask

    def apply_mask(self):
        # Convert binary mask to 3 channels
        binary_mask_3c = cv2.cvtColor(self.binary_mask, cv2.COLOR_GRAY2BGR)
        # Apply the mask to create the result image
        result_image = np.where(binary_mask_3c == 255, self.obj_image, self.bg2_image)
        return result_image

    def display_image(self):
        self.display_image_process(self.obj_image, "Object Image")
        self.display_image_process(self.bg1_image, "Green Screen Background")
        self.display_image_process(self.bg2_image, "New Background Image")
        self.display_image_process(self.result_image, "Result Image")

    def display_image_process(self, image, title="Image"):
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        plt.figure(figsize=(10, 5))
        plt.imshow(image_rgb)
        plt.title(title)
        plt.axis('off')
        plt.show()

image_path = './content/Week-2-Module-2/Object.png'
new_background_path = './content/Week-2-Module-2/NewBackground.jpg'

module = replace_background_module(image_path, new_background_path)
result_image = module.result_image
module.display_image()
```
